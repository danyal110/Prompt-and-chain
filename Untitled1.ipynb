```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.exceptions import OutputParserException
```

```python
promt='''
From the below news article extract about prime minister in JSON format containing following
keys:"Prime Minister","Political Party".

Only return valid JSON.No preamble

Article
===
{article}
'''
pt=PromptTemplate.from_template(promt)
pt
```

```python
import getpass
import os

if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API key: ")
```

```python
llm=ChatGroq(model="llama-3.3-70b-versatile")
chain=pt | llm
```

```python
article_text='''
In a historic shift for the nation’s political trajectory, Tarique Rahman is set to be sworn in as the Prime Minister of Bangladesh on Tuesday afternoon, February 17. Earlier today, Rahman, along with all elected BNP MPs, took oath at the National Parliament building in Dhaka. This landmark ceremony follows the 13th Parliamentary Elections held on February 12, where the Bangladesh Nationalist Party (BNP) secur
'''
res=chain.invoke({"article":article_text})
```

```python
res.content
```

```python
parser=JsonOutputParser()
output=parser.parse(res.content)
```
